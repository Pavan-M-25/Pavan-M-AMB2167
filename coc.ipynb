```python
# @title
print("--- Crop Cost of Cultivation Calculator ---")

    # 1. Input basic details
crop_name = input("Enter crop name: ")
area_size = float(input("Enter area size (in acres): "))

    # 2. Variable Costs (Inputs and Labor)
print("\n--- Variable Costs ---")
seeds = float(input("Seed cost: "))
fertilizers = float(input("Fertilizers & Pesticides: "))
labor = float(input("Labor costs (Sowing, Weeding, Harvesting): "))
machinery = float(input("Machinery hire (Tractor/Irrigation): "))
misc_variable = float(input("Other variable costs: "))

    # 3. Fixed Costs
print("\n--- Fixed Costs ---")
land_rent = float(input("Land rent (or opportunity cost): "))
interest_on_capital = float(input("Interest on working capital (%): ")) / 100

    # Calculations
total_variable_costs = seeds + fertilizers + labor + machinery + misc_variable
interest_amount = total_variable_costs * interest_on_capital
total_cost = total_variable_costs + interest_amount + land_rent
cost_per_acre = total_cost / area_size

# Display Results
print(f"\n{'='*30}")
print(f"COST SUMMARY FOR: {crop_name.upper()}")
print(f"{'='*30}")
print(f"Total Variable Costs:  ${total_variable_costs:,.2f}")
print(f"Interest Applied:      ${interest_amount:,.2f}")
print(f"Fixed Costs:           ${land_rent:,.2f}")
print(f"---")
print(f"TOTAL COST:            ${total_cost:,.2f}")
print(f"COST PER ACRE:         ${cost_per_acre:,.2f}")
print(f"{'='*30}")
```
